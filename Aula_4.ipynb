```python
import requests
import pandas as pd

# URL da API do GitHub para listar o conteúdo do diretório
api_url = 'https://api.github.com/repos/rodolfoharley/7_Days_of_Code_Alura-Python-Pandas/contents/Dia_1-Importando_dados/Datasets/dados_emprestimos'

response = requests.get(api_url)
files = response.json()

# Filtrar apenas os arquivos CSV
csv_files = [file['download_url'] for file in files if file['name'].endswith('.csv')]

# Lista para armazenar os dataframes
dfs = []

for file_url in csv_files:
    df = pd.read_csv(file_url)
    dfs.append(df)

# Concatenar todos os dataframes
emprestimos = pd.concat(dfs, ignore_index=True)

# Exibir o dataframe final
emprestimos.tail()
```

```python
# Verifica se há valores repetidos
emprestimos.value_counts()
```

```python
#apaga valores repetidos
emprestimos = emprestimos.drop_duplicates()
```

```python
import requests
import pandas as pd

# URL do arquivo .parquet no GitHub
url = 'https://github.com/rodolfoharley/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet'

# Fazer o download do arquivo
response = requests.get(url)
with open('temp.parquet', 'wb') as f:
    f.write(response.content)

# Ler o arquivo .parquet baixado
acervo = pd.read_parquet('temp.parquet', engine='pyarrow')

# Exibir as primeiras linhas do dataframe
acervo.head()
```

```python
acervo.value_counts()
```

```python
emprestimos_X_acervo = pd.merge(emprestimos, acervo, on='codigo_barras', how='inner')
emprestimos_X_acervo.dropna()
```

```python
# Criando o dataframe com CDU - Classificação Decimal Universal
data = {
    'range': ['000 a 099', '100 a 199', '200 a 299', '300 a 399', '400 a 499',
              '500 a 599', '600 a 699', '700 a 799', '800 a 899', '900 a 999'],
    'categoria': ['Generalidades. Ciência e conhecimento.', 'Filosofia e psicologia.',
                  'Religião.', 'Ciências sociais.', 'Classe vaga. Provisoriamente não ocupada.',
                  'Matemática e ciências naturais.', 'Ciências aplicadas.', 'Belas artes.',
                  'Linguagem. Língua. Linguística.', 'Geografia. Biografia. História.']
}

cdu = pd.DataFrame(data)

# Separando os limites inferiores e superiores do intervalo
cdu[['limite_inferior', 'limite_superior']] = cdu['range'].str.split(' a ', expand=True)
cdu['limite_inferior'] = cdu['limite_inferior'].astype(int)
cdu['limite_superior'] = cdu['limite_superior'].astype(int)

cdu
```

```python
# Garantir que os dataframes estão ordenados
cdu = cdu.sort_values('limite_inferior')
emprestimos_X_acervo = emprestimos_X_acervo.sort_values('localizacao')
```

```python
# Usando merge_asof para combinar os dataframes com base nos limites
emprestimos_X_acervo_X_cdu = pd.merge_asof(emprestimos_X_acervo, cdu, left_on='localizacao', right_on='limite_inferior', direction='backward')
emprestimos_X_acervo_X_cdu = emprestimos_X_acervo_X_cdu.drop(['range','limite_inferior','limite_inferior','limite_superior','registro_sistema'], axis=1)
emprestimos_X_acervo_X_cdu['matricula_ou_siape'] = emprestimos_X_acervo_X_cdu['matricula_ou_siape'].astype(str)
emprestimos_X_acervo_X_cdu.head()
```

```python
# Transformando string em datetime
emprestimos_X_acervo_X_cdu['data_emprestimo'] = pd.to_datetime(emprestimos_X_acervo_X_cdu['data_emprestimo'])
emprestimos_X_acervo_X_cdu.head()
```

```python
# Agrupando por Ano
emprestimos_X_acervo_X_cdu['ano_emprestimo'] = emprestimos_X_acervo_X_cdu['data_emprestimo'].dt.year
emprestimos_X_acervo_X_cdu.groupby(['ano_emprestimo', 'id_exemplar']).size().reset_index(name='count')

# Agrupando por ano e contando o número de id_exemplar distintos
result_df = emprestimos_X_acervo_X_cdu.groupby('ano_emprestimo')['id_exemplar'].nunique().reset_index(name='count').sort_values('ano_emprestimo')
result_df
```

```python
import pandas as pd
import plotly.express as px

# Supondo que 'emprestimos_X_acervo_X_cdu' já está carregado com os dados e a coluna 'data_emprestimo' está no formato datetime

# Criando o gráfico interativo
fig = px.line(
    result_df,
    x='ano_emprestimo',
    y='count',
    markers=True,
    labels={'ano_emprestimo': 'Ano', 'count': 'Número de Exemplares Distintos'},
    title='Número de Exemplares Distintos por Ano'
)

# Definir cores acessíveis
fig.update_traces(line_color='#377eb8', marker_color='#e41a1c')  # Azul e Vermelho adaptados para CUD

# Alterar o tema do gráfico para maior contraste
fig.update_layout(template="plotly_white")


# Exibindo o gráfico
fig.show()
```

```python
# Agrupando por Ano
emprestimos_X_acervo_X_cdu['mes_emprestimo'] = emprestimos_X_acervo_X_cdu['data_emprestimo'].dt.month_name()
emprestimos_X_acervo_X_cdu.groupby(['mes_emprestimo', 'id_exemplar']).size().reset_index(name='count')

# Agrupando por ano e contando o número de id_exemplar distintos
result_mes = emprestimos_X_acervo_X_cdu.groupby('mes_emprestimo')['id_exemplar'].nunique().reset_index(name='count').sort_values('count', ascending=True)
result_mes
```

```python
import pandas as pd
import plotly.express as px


# Criar um mapeamento de meses para números
month_mapping = {
    'Janeiro': 1, 'Fevereiro': 2, 'Março': 3, 'Abril': 4, 'Maio': 5, 'Junho': 6,
    'Julho': 7, 'Agosto': 8, 'Setembro': 9, 'Outubro': 10, 'Novembro': 11, 'Dezembro': 12
}

# Adicionar uma coluna com números dos meses
result_mes['mes_numero'] = result_mes['mes_emprestimo'].map(month_mapping)

# Ordenar o dataframe pelos números dos meses
result_mes = result_mes.sort_values('mes_numero')

# Criar o gráfico interativo
fig = px.line(
    result_mes,
    x='mes_emprestimo',
    y='count',
    markers=True,
    labels={'mes_emprestimo': 'Mês', 'count': 'Total de empréstomos de exemplares'},
    title='Total de Empréstimos de Exemplares por Mês'
)

# Definir cores acessíveis
fig.update_traces(line_color='#377eb8', marker_color='#e41a1c')  # Azul e Vermelho adaptados para CUD

# Alterar o tema do gráfico para maior contraste
fig.update_layout(template="plotly_white")

# Exibir o gráfico
fig.show()

```

```python
# Agrupando por Ano
emprestimos_X_acervo_X_cdu['hora_emprestimo'] = emprestimos_X_acervo_X_cdu['data_emprestimo'].dt.hour
emprestimos_X_acervo_X_cdu.groupby(['hora_emprestimo', 'id_exemplar']).size().reset_index(name='count')

# Agrupando por ano e contando o número de id_exemplar distintos
result_hora = emprestimos_X_acervo_X_cdu.groupby('hora_emprestimo')['id_exemplar'].nunique().reset_index(name='count').sort_values('count', ascending=True)
result_hora
```

```python
import pandas as pd

def gerar_tabela_frequencia(df, coluna):
    # Contagem de frequência
    frequencia = df[coluna].value_counts().reset_index()
    frequencia.columns = [coluna, 'Frequencia']

    # Cálculo do percentual
    frequencia['Percentual'] = round((frequencia['Frequencia'] / frequencia['Frequencia'].sum()) * 100,2)

    return frequencia

# Exemplo de uso:
tabela_tipo_vinculo = gerar_tabela_frequencia(emprestimos_X_acervo_X_cdu, 'tipo_vinculo_usuario')
tabela_colecao = gerar_tabela_frequencia(emprestimos_X_acervo_X_cdu, 'colecao')
tabela_biblioteca = gerar_tabela_frequencia(emprestimos_X_acervo_X_cdu, 'biblioteca')
tabela_categoria = gerar_tabela_frequencia(emprestimos_X_acervo_X_cdu, 'categoria')


# Exibindo o resultado
print(tabela_tipo_vinculo)
print(tabela_colecao)
print(tabela_biblioteca)
print(tabela_categoria)
```
