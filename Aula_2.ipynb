```python
import requests
import pandas as pd

# URL da API do GitHub para listar o conteúdo do diretório
api_url = 'https://api.github.com/repos/rodolfoharley/7_Days_of_Code_Alura-Python-Pandas/contents/Dia_1-Importando_dados/Datasets/dados_emprestimos'

response = requests.get(api_url)
files = response.json()

# Filtrar apenas os arquivos CSV
csv_files = [file['download_url'] for file in files if file['name'].endswith('.csv')]

# Lista para armazenar os dataframes
dfs = []

for file_url in csv_files:
    df = pd.read_csv(file_url)
    dfs.append(df)

# Concatenar todos os dataframes
emprestimos = pd.concat(dfs, ignore_index=True)

# Exibir o dataframe final
emprestimos.tail()
```

```python
# Verifica se há valores repetidos
emprestimos.value_counts()
```

```python
#apaga valores repetidos
emprestimos = emprestimos.drop_duplicates()
```

```python
import requests
import pandas as pd

# URL do arquivo .parquet no GitHub
url = 'https://github.com/rodolfoharley/7_Days_of_Code_Alura-Python-Pandas/raw/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet'

# Fazer o download do arquivo
response = requests.get(url)
with open('temp.parquet', 'wb') as f:
    f.write(response.content)

# Ler o arquivo .parquet baixado
acervo = pd.read_parquet('temp.parquet', engine='pyarrow')

# Exibir as primeiras linhas do dataframe
acervo.head()
```

```python
acervo.value_counts()
```

```python
emprestimos_X_acervo = pd.merge(emprestimos, acervo, on='codigo_barras', how='inner')
emprestimos_X_acervo.dropna()
```

```python
# Criando o dataframe com CDU - Classificação Decimal Universal
data = {
    'range': ['000 a 099', '100 a 199', '200 a 299', '300 a 399', '400 a 499',
              '500 a 599', '600 a 699', '700 a 799', '800 a 899', '900 a 999'],
    'categoria': ['Generalidades. Ciência e conhecimento.', 'Filosofia e psicologia.',
                  'Religião.', 'Ciências sociais.', 'Classe vaga. Provisoriamente não ocupada.',
                  'Matemática e ciências naturais.', 'Ciências aplicadas.', 'Belas artes.',
                  'Linguagem. Língua. Linguística.', 'Geografia. Biografia. História.']
}

cdu = pd.DataFrame(data)

# Separando os limites inferiores e superiores do intervalo
cdu[['limite_inferior', 'limite_superior']] = cdu['range'].str.split(' a ', expand=True)
cdu['limite_inferior'] = cdu['limite_inferior'].astype(int)
cdu['limite_superior'] = cdu['limite_superior'].astype(int)

cdu
```

```python
# Garantir que os dataframes estão ordenados
cdu = cdu.sort_values('limite_inferior')
emprestimos_X_acervo = emprestimos_X_acervo.sort_values('localizacao')
```

```python
# Usando merge_asof para combinar os dataframes com base nos limites
emprestimos_X_acervo_X_cdu = pd.merge_asof(emprestimos_X_acervo, cdu, left_on='localizacao', right_on='limite_inferior', direction='backward')
emprestimos_X_acervo_X_cdu = emprestimos_X_acervo_X_cdu.drop(['range','limite_inferior','limite_inferior','limite_superior','registro_sistema'], axis=1)
emprestimos_X_acervo_X_cdu['matricula_ou_siape'] = emprestimos_X_acervo_X_cdu['matricula_ou_siape'].astype(str)
emprestimos_X_acervo_X_cdu.head()
```
